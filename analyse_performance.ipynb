```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculation of the correlation matrix # Calcul de la matrice de corrlation
corr_matrix = data.corr(numeric_only=True)

# Displaying the heatmap # Affichage de la heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de Corrlation entre les Variables Numriques")
plt.show()
```

```python
# Performance score by department # Score de performance par dpartement
plt.figure(figsize=(12, 6))
sns.boxplot(x="Department", y="Performance_Score", data=data)
plt.xticks(rotation=45)
plt.title("Score de Performance par Dpartement")
plt.show()

# Performance score by education level # Score de performance par niveau d'ducation
plt.figure(figsize=(10, 5))
sns.boxplot(x="Education_Level", y="Performance_Score", data=data)
plt.xticks(rotation=45)
plt.title("Score de Performance par Niveau d'ducation")
plt.show()
```

```python
# Calculation of the average salary by department # Calcul du salaire moyen par dpartement
salary_by_department = data.groupby("Department")["Monthly_Salary"].mean().sort_values()

# Displaying the chart # Affichage du graphique
plt.figure(figsize=(12, 6))
sns.barplot(x=salary_by_department.index, y=salary_by_department.values)
plt.xticks(rotation=45)
plt.title("Salaire Moyen par Dpartement")
plt.ylabel("Salaire Moyen")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data["Projects_Handled"], y=data["Performance_Score"])
plt.title("Relation entre le Nombre de Projets et le Score de Performance")
plt.xlabel("Projects_Handled")
plt.ylabel("Performance_Score")
plt.show()
```

```python
# Interprtation et prochaines tapes
# Avec ces analyses, nous avons pu identifier plusieurs tendances :
# Les projets grs et les heures de formation influencent positivement la performance.
# Les heures supplmentaires et les absences semblent nuire  la satisfaction des employs.
# Les carts de salaire varient fortement selon le dpartement.
```

```python
# Interpretation
# With these analyses, we have identified several trends:

# Managed projects and training hours positively influence performance.
# Overtime and absences seem to impact employee satisfaction negatively.
# Salary discrepancies vary significantly across departments.
```
